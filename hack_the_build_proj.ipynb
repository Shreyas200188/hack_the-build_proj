<h2>Train test split</h2>

```
from sklearn.model_selection import train_test_split
```

```
X = combined_raw_df.drop("prognosis",axis =1)
Y = combined_raw_df["prognosis"]
```

```
X,Y
```

```
X_train , X_test , Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```
X_train,Y_train
```

<strong>Neural Network </strong>

```
values=np.unique(Y_train)
values
```

```
Y_train_2 = np.zeros(shape=(len(Y_train),len(values)))
k=0
for x in Y_train:
     for i in range(len(values)):
            if x==values[i]:
                tmp=list(np.zeros(41))
                tmp[i]=1
                Y_train_2[k]=tmp
                k+=1
```

```
Y_train_2
```

```
model = Sequential()
model.add(Dense(3969, input_dim=128,activation='tanh'))
model.add(Dense(41,activation='sigmoid'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy',metrics=['accuracy'], optimizer=sgd)
model.fit(X_train, Y_train_2, batch_size=1, epochs=10)
```

```
test_values=np.unique(Y_test)
Y_test_2 = np.zeros(shape=(len(Y_test),len(test_values)))
k=0
for x in Y_test:
     for i in range(len(test_values)):
            if x==test_values[i]:
                tmp=list(np.zeros(41))
                tmp[i]=1
                Y_test_2[k]=tmp
                k+=1
```

```
pre=model.predict_proba(X_test)
```

```
acc=0;
for i in range(len(pre)):
    if pre[i].argmax() == Y_test_2[i].argmax() :
        if pre[i].max() >= 0.90:
            acc+=1

acc=acc/len(pre)
acc
```

<h3>Random Forest</h3>

```
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500,max_depth =9,random_state=0,criterion='entropy')
```

```
rf.fit(X_train,Y_train)
```

```
rf_train_pred = rf.predict(X_train)
rf_train_pred
```

```
(Y_train==rf_train_pred).sum()/len(Y_train)
```

```
rf_test_pred = rf.predict(X_test)
```

```
(Y_test==rf_test_pred).sum()/len(Y_test)
```

<h3>Decsion Tree</h3>

```
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.tree import DecisionTreeClassifier           #building the model
dt = DecisionTreeClassifier(random_state=0,max_depth = 50)
dt.fit(X_train,Y_train)
dt_pred_train = dt.predict(X_train)            #predicting on training values
dt_pred_test  =dt.predict(X_test)              #predicting on test values
print("Test Accuracy : ",(Y_test==dt_pred_test).sum()/len(Y_test))
print("Train Accuracy : ",(Y_train==dt_pred_train).sum()/len(Y_train))
```

```
accuracy_score(Y_train,dt_pred_train)
```

<h3>SVM</h3>

```
from sklearn import svm
svc = svm.SVC(kernel='rbf')
svc.fit(X_train,Y_train)
svc_pred_test = svc.predict(X_test)
svc_pred_train = svc.predict(X_train)
```

```
print("Training Accuracy : ",accuracy_score(Y_train,svc_pred_train))
print("Test Accuracy : ",accuracy_score(Y_test,svc_pred_test))
```
